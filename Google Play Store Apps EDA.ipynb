```python
# Cell 1 - Import Libraries and Initial Setup

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings


# Ignore Warnings
warnings.filterwarnings('ignore')

# Set plot styles for consistency
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)
plt.rcParams['axes.titlesize'] = 14
plt.rcParams['axes.labelsize'] = 12

# Pandas display options for better readability
pd.set_option('display.max_columns', 100)
pd.set_option('display.width', 120)

# Suppress scientific notation for clarity
pd.set_option('display.float_format', lambda x: '%.3f' % x)

print("Libraries imported and settings configured.")
```

```python
# Cell 2 - Load Data and Initial Exploration

# Load the dataset
df = pd.read_csv('/kaggle/input/google-playstore-dataset/google_playstore.csv')

# Show the shape of the dataframe
print(f'Dataset shape: {df.shape}')

# Display the first 5 rows
display(df.head())

# Display basic information about columns, datatypes, missing values
print("\nDataFrame Info:")
df.info()

# Show summary statistics for numeric columns
print("\nSummary Statistics for Numeric Columns:")
display(df.describe())

# Show summary statistics for object (categorical) columns
print("\nSummary Statistics for Object Columns:")
display(df.describe(include=['object']))
```

```python
# Cell 3 - Data Quality Checks & Unique Values

# Missing values per column
print("Missing values per column:")
display(df.isnull().sum().sort_values(ascending=False))

# Duplicates
num_duplicates = df.duplicated().sum()
print(f"\nNumber of duplicate rows: {num_duplicates}")

# Unique values per column
print("\nUnique values per column:")
display(df.nunique().sort_values(ascending=False))
```

```python
# Cell 4 - Data Cleaning: Type Conversion & Handling Missing Values

# Convert 'Released' and 'Last Updated' to datetime
df['Released'] = pd.to_datetime(df['Released'], errors='coerce')
df['Last Updated'] = pd.to_datetime(df['Last Updated'], errors='coerce')
df['Scraped Time'] = pd.to_datetime(df['Scraped Time'], errors='coerce')

# Clean 'Installs' and 'Minimum Installs' columns: remove commas, plus signs, convert to float
for col in ['Installs', 'Minimum Installs']:
    df[col] = df[col].replace('[+,]', '', regex=True).astype(float)

# Clean 'Size' column: convert to MB where possible, handle 'Varies with device'
import re
def parse_size(x):
    if pd.isnull(x) or x == 'Varies with device':
        return np.nan
    match = re.match(r'([0-9\.]+)([kKmMgG])', x)
    if match:
        size = float(match.group(1))
        unit = match.group(2).upper()
        if unit == 'K':
            return size / 1024
        elif unit == 'M':
            return size
        elif unit == 'G':
            return size * 1024
    return np.nan

df['Size_MB'] = df['Size'].map(parse_size)

# Convert 'Price' to float if not already
df['Price'] = df['Price'].astype(float)

# Print confirmation and a summary of the cleaned columns
print("Data types after conversion:")
print(df[['Released', 'Last Updated', 'Scraped Time', 'Installs', 'Minimum Installs', 'Size', 'Size_MB', 'Price']].dtypes)

print("\nSample cleaned data (first 5 rows):")
display(df[['Installs', 'Minimum Installs', 'Size', 'Size_MB', 'Price']].head())

print("\nMissing values after cleaning (relevant columns):")
display(df[['Size', 'Size_MB', 'Installs', 'Minimum Installs', 'Released', 'Last Updated']].isnull().sum())
```

```python
# Cell 5 - Missing Value Imputation & Dataset Consistency

# Impute or handle missing numeric values
# For 'Size_MB', impute with median (common practice for skewed app size distribution)
size_mb_median = df['Size_MB'].median()
df['Size_MB'].fillna(size_mb_median, inplace=True)

# For 'Installs' and 'Minimum Installs', impute with 0 (assume apps with missing installs are very new or rare)
df['Installs'].fillna(0, inplace=True)
df['Minimum Installs'].fillna(0, inplace=True)

# For 'Released', we can leave missing as NaT, or optionally drop those rows if required for time-based analysis

# Show summary of missing values after imputation
print("Missing values after imputation:")
display(df[['Size_MB', 'Installs', 'Minimum Installs', 'Released']].isnull().sum())

# Final dataset shape
print(f"\nFinal dataset shape: {df.shape}")

# Preview a random sample of 5 rows to confirm cleaning
display(df.sample(5))
```

```python
# Cell 6 - Univariate Analysis: Visual & Statistical Summaries

# Numeric columns to plot
num_cols = ['Rating', 'Rating Count', 'Installs', 'Size_MB', 'Price']

# Plot numeric columns
for col in num_cols:
    plt.figure(figsize=(8, 4))
    sns.histplot(df[col], bins=50, kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()

    print(f"\nSummary statistics for {col}:")
    display(df[col].describe())

# Categorical columns to plot (top 10 most common for readability)
cat_cols = ['Category', 'Content Rating', 'Free']

for col in cat_cols:
    plt.figure(figsize=(10, 4))
    if df[col].dtype == 'bool':
        sns.countplot(x=col, data=df)
    else:
        top_cats = df[col].value_counts().nlargest(10)
        sns.barplot(x=top_cats.index, y=top_cats.values)
        plt.xticks(rotation=45)
    plt.title(f'Distribution of {col}')
    plt.ylabel('Count')
    plt.xlabel(col)
    plt.show()

    print(f"\nTop 10 counts for {col}:")
    display(df[col].value_counts().head(10))
```

```python
# Cell 7 - FAST Bivariate Analysis & Correlation (for large datasets)

# 1. Correlation heatmap (on full data, since it's only a summary)
numeric_cols = ['Rating', 'Rating Count', 'Installs', 'Minimum Installs', 'Maximum Installs', 'Size_MB', 'Price']
corr = df[numeric_cols].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()

# 2. Pairplot (on a much smaller sample, e.g., 1,000 rows)
sample_df = df[numeric_cols].dropna().sample(1000, random_state=42)
sns.pairplot(sample_df)
plt.suptitle('Pairplot of Main Numeric Features (Sample of 1,000)', y=1.02)
plt.show()

# 3. Boxplot: Rating by Category (top 8 categories only, sample 5,000 rows)
top_cats = df['Category'].value_counts().nlargest(8).index
cat_sample = df[df['Category'].isin(top_cats)].sample(5000, random_state=42)
plt.figure(figsize=(12, 6))
sns.boxplot(x='Category', y='Rating', data=cat_sample)
plt.xticks(rotation=45)
plt.title('App Rating by Top 8 Categories (Sample)')
plt.show()

# 4. Boxplot: Rating by Free vs Paid (sample 5,000)
free_sample = df.sample(5000, random_state=42)
plt.figure(figsize=(6, 4))
sns.boxplot(x='Free', y='Rating', data=free_sample)
plt.title('App Rating: Free vs Paid (Sample)')
plt.show()

# 5. Scatter plot: Installs vs Rating (sample 5,000, log scale)
scatter_sample = df[['Installs', 'Rating']].dropna().sample(5000, random_state=42)
plt.figure(figsize=(8, 4))
sns.scatterplot(x=np.log1p(scatter_sample['Installs']), y=scatter_sample['Rating'], alpha=0.2)
plt.xlabel('Log(Installs + 1)')
plt.ylabel('Rating')
plt.title('Rating vs. Installs (Log Scale, Sample)')
plt.show()

print('Cell 7 successfully executed')
```

```python
# Cell 8 - Multivariate Analysis & Feature Engineering Ideas

# 1. Pivot table: Average Rating by Category and Free/Paid
pivot = df.pivot_table(values="Rating", index="Category", columns="Free", aggfunc="mean")
plt.figure(figsize=(10, 6))
sns.heatmap(pivot, annot=True, fmt=".2f", cmap="coolwarm")
plt.title('Average App Rating by Category & Free/Paid')
plt.ylabel('Category')
plt.xlabel('Is Free')
plt.show()

# 2. Price vs. Installs for paid apps (log scale for price and installs)
paid = df[df['Free'] == False]
plt.figure(figsize=(8, 4))
sns.scatterplot(x=np.log1p(paid['Price']), y=np.log1p(paid['Installs']), alpha=0.2)
plt.xlabel('Log(Price + 1)')
plt.ylabel('Log(Installs + 1)')
plt.title('Price vs Installs for Paid Apps')
plt.show()

# 3. Boxplots for outlier detection in key numeric columns
num_cols = ['Rating', 'Rating Count', 'Installs', 'Size_MB', 'Price']
for col in num_cols:
    plt.figure(figsize=(8, 1.5))
    sns.boxplot(x=df[col], orient='h')
    plt.title(f'Outlier Check: {col}')
    plt.show()
```

```python
# Cell 9
# Helper function to remove outliers using IQR method
def remove_outliers_iqr(df, cols):
    df_clean = df.copy()
    for col in cols:
        if df_clean[col].dtype != 'O':  # skip object columns
            Q1 = df_clean[col].quantile(0.25)
            Q3 = df_clean[col].quantile(0.75)
            IQR = Q3 - Q1
            lower = Q1 - 1.5 * IQR
            upper = Q3 + 1.5 * IQR
            df_clean = df_clean[(df_clean[col] >= lower) & (df_clean[col] <= upper)]
    return df_clean

# Columns to treat for outliers
num_cols = ['Rating', 'Rating Count', 'Installs', 'Size_MB', 'Price']

# Remove outliers
df_no_outliers = remove_outliers_iqr(df, num_cols)

print(f"Original shape: {df.shape}")
print(f"Shape after removing outliers: {df_no_outliers.shape}")

# EDA on Cleaned Data

# Histograms
for col in num_cols:
    plt.figure(figsize=(7,3))
    sns.histplot(df_no_outliers[col], bins=40, kde=True)
    plt.title(f'No-Outlier Distribution: {col}')
    plt.show()

# Boxplots
for col in num_cols:
    plt.figure(figsize=(7,1.5))
    sns.boxplot(x=df_no_outliers[col], orient='h')
    plt.title(f'No-Outlier Boxplot: {col}')
    plt.show()

# Correlation Heatmap
plt.figure(figsize=(8,6))
sns.heatmap(df_no_outliers[num_cols].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap (No Outliers)')
plt.show()
```
